```python
from PyQt6.QtWidgets import QMainWindow, QLineEdit, QApplication, QPushButton, QVBoxLayout, QWidget, QHBoxLayout
from PyQt6.QtCore import Qt
from PyQt6.QtGui import QIcon
import sys

class MyWindow(QMainWindow):
    def __init__(self):
        super().__init__()
        self.initUI()
    
    def initUI(self):
        central_widget = QWidget()
        self.setCentralWidget(central_widget)
        layout = QVBoxLayout(central_widget)
        
        # Create QLineEdit widgets
        self.edt_mk_cu = QLineEdit(self)
        self.edt_mk_cu.setEchoMode(QLineEdit.EchoMode.Password)  # Hide input with password mode
        
        self.edt_mk_moi = QLineEdit(self)
        self.edt_mk_moi.setEchoMode(QLineEdit.EchoMode.Password)  # Hide input with password mode
        
        # Create buttons for show/hide password
        self.btn_toggle_password_cu = QPushButton(self)
        self.btn_toggle_password_cu.setIcon(QIcon("eye_closed_icon.png"))
        self.btn_toggle_password_cu.setCheckable(True)
        self.btn_toggle_password_cu.clicked.connect(lambda: self.toggle_password_visibility(self.edt_mk_cu, self.btn_toggle_password_cu))
        
        self.btn_toggle_password_moi = QPushButton(self)
        self.btn_toggle_password_moi.setIcon(QIcon("eye_closed_icon.png"))
        self.btn_toggle_password_moi.setCheckable(True)
        self.btn_toggle_password_moi.clicked.connect(lambda: self.toggle_password_visibility(self.edt_mk_moi, self.btn_toggle_password_moi))
        
        # Layout for the old password field and button
        layout_cu = QHBoxLayout()
        layout_cu.addWidget(self.edt_mk_cu)
        layout_cu.addWidget(self.btn_toggle_password_cu)
        
        # Layout for the new password field and button
        layout_moi = QHBoxLayout()
        layout_moi.addWidget(self.edt_mk_moi)
        layout_moi.addWidget(self.btn_toggle_password_moi)
        
        # Add widgets to main layout
        layout.addLayout(layout_cu)
        layout.addLayout(layout_moi)
        
        self.setGeometry(300, 300, 300, 200)
        self.setWindowTitle('Password Input Example')

    def toggle_password_visibility(self, line_edit, button):
        if button.isChecked():
            line_edit.setEchoMode(QLineEdit.EchoMode.Normal)
            button.setIcon(QIcon("eye_open_icon.png"))
        else:
            line_edit.setEchoMode(QLineEdit.EchoMode.Password)
            button.setIcon(QIcon("eye_closed_icon.png"))

    def get_input(self):
        input_mk_cu = self.edt_mk_cu.text()
        input_mk_moi = self.edt_mk_moi.text()
        print(f"Old Password: {input_mk_cu}")
        print(f"New Password: {input_mk_moi}")

if __name__ == '__main__':
    app = QApplication(sys.argv)
    window = MyWindow()
    window.show()
    sys.exit(app.exec())
```
